```python
import pandas as pd
import numpy as np
```

```python
tit_train = pd.read_csv("/Users/garrettbrown/Desktop/CS189/HWs/hw5_starter/datasets/titanic/titanic_training.csv")
tit_train
```

```python
#ABC seem to 1st class, DEF 2nd class, , EFG 3rd class
tit_train[tit_train["pclass"] == 3]["cabin"].value_counts()
```

```python
tit_train.isna().sum()
```

```python
#drop the one unlabeled training datum
tit_train = tit_train.dropna(subset = ["survived"])
tit_train.isna().sum()
```

```python
#we will impute the one missing fare by class:
median = tit_train[tit_train["pclass"] == 3]["fare"].median()
tit_train.loc[:, "fare"] = tit_train["fare"].fillna(median)
tit_train.isna().sum()
```

```python
#bin missing embarcation as U
tit_train.loc[:, "embarked"] = tit_train["embarked"].fillna("U")
tit_train.isna().sum()
```

```python
#To simplify the model for now, drop the ticket number columns
tit_train = tit_train.drop(columns = ["ticket"])
```

```python
#bin the ages
bins = [0, 5, 13, 21, 35, 60, 100]
labels = ["child", "preteen", "teen", "young adult", "adult", "elderly"]
tit_train["age group"] = pd.cut(tit_train["age"], bins = bins, labels = labels)
tit_train = tit_train.drop(columns = ["age"])
tit_train["age group"] = tit_train["age group"].astype("str")
tit_train["age group"] = tit_train["age group"].replace("nan", "N")
```

```python
#just only look at the letter of the cabin
tit_train["cabin"] = tit_train["cabin"].str[0]
tit_train["cabin"] = tit_train["cabin"].fillna("N")
```

```python
#cleaned data
tit_train
```

```python
tit_train.to_csv('/Users/garrettbrown/Desktop/CS189/HWs/hw5_starter/titanic_train_cleaned.csv', index=False)
```
